```python
import csv
import urllib.request
from bs4 import BeautifulSoup
```

```python
def get_html(url):
    response = urllib.request.urlopen(url)
    return response.read()

def get_page_count(html):
    soup = BeautifulSoup(html)
    pagginator = soup.find('div', class_='col-1 col-sm-2 text-right').find('a')
    pagginator = int(str(pagginator)[21:24])
    
    
    return pagginator

def parse(html):
    soup = BeautifulSoup(html)
    table = soup.find_all('div', class_='cols_table')[2]

    projects = []

    for row in table.find_all('div', class_='row'):
        #cols = row.find_all('a', class_='text-bold show_visited')
        #print(cols)
        categ = row.find_all('div')[4].find_all('a')[:]
        categ = [category.text.replace('\xa0', ' ') for category in row.find_all('div')[4].find_all('a')[:]]
        #print(categ)
        
        projects.append({
            'title': row.a.text,
            'price': row.find_all('div')[2].text.strip(),
            'application': row.find_all('div')[3].text.strip().replace(' заявок', '').replace(' заявка', '').replace(' заявки', ''),
            'categories': categ
        })
        
    #for project in projects:
        #print(projects)
        
    #rows = table.find_all('div', class_='row')
    #print(rows)
    #logo = table.find('a', class_='text-bold show_visited')
    #print(logo)
    
    return projects
    
def save(projects, path):
    with open(path, 'w') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(('Проект', 'Цена', 'Заявки', 'Категории'))
        
        for project in projects:
            writer.writerow((project['title'], project['price'], project['application'], ', '.join(project['categories'])))
    
    
    
def main():
    #page_count = get_page_count('http://weblancer.net/projects')
    
    page_count = get_page_count(get_html('http://weblancer.net/projects'))
    
    print('Всего найдено страниц %d' % page_count)
    
    projects = []
    
    for page in range(1, page_count+1):
        print('Парсинг %d%%' % (page / page_count * 100))
        projects.extend(parse(get_html('http://weblancer.net/projects' + '?page=%d' % page)))
            
    #print(parse(get_html('http://weblancer.net/projects')))
    
    save(projects, 'projects.csv')

if __name__ == '__main__':
    main()
```
